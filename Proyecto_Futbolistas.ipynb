```python
#Importar libreria pandas
import pandas as pd

#Accedemos a los datos a traves de un csv
data = pd.read_csv("SalaryPrediction.csv")

#Solicitamos informacion del csv
data.info()
```

```python
#Exploramos los datos de inicio
data.head()
```

```python
#Exploramos los ultimos datos
data.tail()
```

```python
#Descripcion general
data.describe()
```

```python
#Tipos de datos de cada columna
data.dtypes
```

```python
#Solicitamos el numero de elementos del data frame
data.size
```

```python
#Eliminar solo las filas con datos desconocidos de la columna 'Wage' y 'Age'
data_limpio = data.dropna(subset=['Wage', 'Age'])
data_limpio
```

```python
#Solicitamos la cantidad de datos del data frame para comparar con el original y comprobar si faltaba algun dato
data_limpio.size
#Como podemos observar en el resultado, los datos de las edades y salarios estan completos, ya que no se han
#eliminado elementos, por lo que podemos seguir trabajando con nuestro 'data' original
```

```python
#Para convertir la columna 'Wage' a float, primero debemos eliminar las comas innecesarias que se encuentran en los valores
data['Wage'] = data['Wage'].replace({',':''}, regex=True)

#Imprimimos la columna 'Wage' para verificar que los datos fueron modificados con éxito
data['Wage']
```

```python
#Convertimos la columna 'Wage' a float
data['Wage'] = data['Wage'].astype(float)

#Para comprobar que la columna fue convertida en float correctamente, solicitamos el tipo de dato
#de dicha columna con el metodo dtypes
data['Wage'].dtypes
```

```python
#Imprimimos la columna wage
print(data['Wage'])
```

```python
#Agrupamos los jugadores por Ligas en las que juegan, y solicitamos la media del salario en cada liga
data.groupby(['League'])['Wage'].mean() 
#Podemos observar como un dato relevante que la liga mejor paga es la Premier League, seguida por la Primeira Liga.
```

```python
#Agrupamos los jugadores por Ligas en las que juegan, y solicitamos la variacion estandar del salario en cada liga
data.groupby(['League'])['Wage'].std() 
#Observamos que la variacion mas grande se da en la Primeira Liga, segundo la Ligue 1 y tercero La Premier League
```

```python
#Agrupamos los jugadores por Ligas en las que juegan, y solicitamos la mediana del salario en cada liga
data.groupby(['League'])['Wage'].median()
#Observamos la mediana es mas grande en Premier League, seguida de la Serie A.
```

```python
#Realizamos analisis a partir de diagramas con Matplotlib y seaborn para evaluar variables y combinaciones
#Importamos libreria matplotlib y seaborn
import seaborn as sns
import matplotlib.pyplot as plt

#Utilizamos un lineplot con colores diferenciados por liga
#para analizar los sueldos en base a la edad
sns.lineplot(x = 'Age', y = 'Wage', hue = 'League', data = data)

#mostrar grafica
plt.show()

#Podemos observar que la Liga mejor paga es la Premier League y generalmente
#Son jugadores entre 26 y 31 años los mejor pagos.
#La Bundesliga por su parte, posee uno o algunos jugadores bien pagos 
#con una edad aproximada de 36 años
#La liga peor paga es la Primeira liga, siendo bastante marcada la diferencia
```

```python
import seaborn as sns
import pandas as pd


#Gráfico box plot para evaluar sueldos en cada liga
sns.boxplot(x = 'League', y = 'Wage', data = data)
plt.xticks(rotation = 90)
#Observamos que hay tres sueldos bastante fuera de rango promedio en
#la Ligue 1
```

```python
#Gráfico box plot para evaluar sueldos segun la posicion del campo
sns.barplot(x = 'Position', y = 'Wage', data = data)
plt.xticks(rotation = 90)
#Podemos observar que quienes mas ganan en promedio suelen ser los 
#mediocampistas('Midfilder'), y quienes menos ganan en promedio son
#los arqueros('Goalkeeper')
```

```python

#Configurar estilo de grafico
sns.set(style = 'whitegrid')

#Ocurrencias de edad
conteo_edades = data['Age'].value_counts().reset_index()

#Renombrar las columnas para mayor claridad
#Veremos la frecuencia de cantidad de jugadores de cada edad
conteo_edades.columns = ['Age', 'Frecuencia']

#Dibujar el bar plot
sns.barplot(x = 'Age', y = 'Frecuencia', data = conteo_edades)
plt.xticks(rotation = 45)#Ajustar la orientacion de las etiquetas del eje x

#Podemos observar que la mayoria de los jugadores son jovenes, siendo
# de 18 y 19 años los que mayor cantidad aportan a las ligas.
```
