```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

#Cargar el archivo csv
data = pd.read_csv("nba.csv")

#Damos una mirada a la informacion de nuestra tabla
print(data.head())
print(data.tail())

#Graficar un lineplot usando las columnas 'Age' y 'Weight'
sns.lineplot(x = 'Edad', y = 'Peso', data = data)
```

```python
#Graficar un lineplot con colores diferenciados por posicion
sns.lineplot(x = 'Edad', y = 'Peso', hue = 'Posicion', data = data)

#mostrar grafica
import matplotlib .pyplot as plt
plt.show()
```

```python
#Graficar un lineplot con colores diferenciados por la posicion
sns.lineplot(x = 'Edad', y = 'Altura', hue = 'Posicion', data = data)
plt.show()
```

```python
#Graficos de barra

import seaborn as sns
import pandas as pd

#Configurar estilo de grafico
sns.set(style = 'whitegrid')

#Leer Csv y contar las frecuencias de cada edad
datos = pd.read_csv("nba.csv")
conteo_edades = data['Edad'].value_counts().reset_index()

#Renombrar las columnas para mayor claridad
#Veremos la frecuencia de cantidad de jugadores de cada edad
conteo_edades.columns = ['Edad', 'Frecuencia']

#Dibujar el bar plot
sns.barplot(x = 'Edad', y = 'Frecuencia', data = conteo_edades)
plt.xticks(rotation = 45)#Ajustar la orientacion de las etiquetas del eje x
```
